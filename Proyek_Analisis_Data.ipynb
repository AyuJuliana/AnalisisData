## Visualization & Explanatory Analysis

### Pertanyaan 1:  Bagaimana pendapatan e-commerce setiap bulannya pada tahun 2017?

```python
# Filter the data to include only the year 2017
orders_2017 = orders_df[orders_df['order_purchase_timestamp'].dt.year == 2017]

# Merge the orders and order_items tables based on order_id
merged_orders_items = pd.merge(orders_2017, order_items_df[['order_id', 'price']], on='order_id')

# Add a column for the month extracted from order_purchase_timestamp
merged_orders_items['purchase_month'] = merged_orders_items['order_purchase_timestamp'].dt.to_period('M')

# Group by month and calculate the total revenue
monthly_revenue_2017 = merged_orders_items.groupby('purchase_month')['price'].sum()

# Displaying the monthly revenue for the year 2017
print(monthly_revenue_2017)
```

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))

# Plot the data with markers and lines
plt.plot(monthly_revenue_2017.index.astype(str), monthly_revenue_2017.values,
         marker='o', linestyle='-', color='skyblue', markersize=8, linewidth=2)

# Set the title and labels
plt.title('Monthly E-commerce Revenue in 2017', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=45)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

## Analisis Lanjutan (Opsional) - RFM Analysis

RFM Analysis, bertujuan mengelompokkan pelanggan berdasarkan perilaku pembelian mereka dengan memperhatikan tiga faktor utama:
1. Recency: Menghitung  kapan terakhir kali customer melakukan transaksi.
2. Frequency: Menghitung jumlah total transaksi/seberapa sering transaksi yang dilakukan oleh pelanggan
3. Monetary: Menghitung total pendapatan perusahaan dari suatu pelanggan

```python
# Merging DataFrames to create a unified dataset
all_df = orders_df.merge(order_items_df, on='order_id') \
                   .merge(order_payments_df, on='order_id') \
                   .merge(customers_df, on='customer_id')

# Calculating RFM values for each unique customer
rfm_df = all_df.groupby('customer_unique_id', as_index=False).agg({
    "order_approved_at": "max",  # Taking the most recent order date
    "order_id": "nunique",       # Counting unique orders (Frequency)
    "price": "sum"               # Summing revenue from the customer (Monetary)
})

# Renaming columns for clarity
rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

# Calculating the last transaction date for customers
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

# Dropping the max_order_timestamp column as it's no longer needed
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

# Displaying the first few rows of the RFM DataFrame
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), hue="customer_unique_id", legend=False, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15, rotation=90)

sns.barplot(y="frequency", x="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), hue="customer_unique_id", legend=False, palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15, rotation=90)

sns.barplot(y="monetary", x="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), hue="customer_unique_id", legend=False, palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15, rotation=90)

plt.suptitle("Customer terbaik berdasarkan parameter RFM (customer_unique_id)", fontsize=20)
plt.show()
```
